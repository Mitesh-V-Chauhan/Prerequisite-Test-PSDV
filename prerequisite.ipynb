```python
def fnct(l1, l2):
    intersection=[]
    union=[]
    symmetric_difference=[]
    for el in l1:
        if el in l2:
            intersection.append(el)
    union= list(set(l1 + l2))
    symmetric_difference=list(set(l1 + l2))
    for elm in union:
        if elm in intersection:
            symmetric_difference.remove(elm)
    print(f'''{intersection}
{union}
{symmetric_difference}''')         
    return
    
fnct([1, 2, 3, 4, 5], [4, 5, 6, 7, 8])
```
